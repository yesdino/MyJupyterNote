    - Bottom-Up

### <a id='Ex1'> Ex.1 Number Problem æ±‚å’Œçš„ä¸åŒæ–¹å¼

Given n, find the number of different ways to write n as the sum of 1, 3, 4.

<a href="#top">å›åˆ°é¡¶éƒ¨</a>

````python
def coin(n):
    dp = [0] * (n + 1)
    dp[0] = dp[1] = dp[2] = 1
    dp[3] = 2
    for i in range(4, n + 1):
        dp[i] = dp[i - 1] + dp[i - 3] + dp[i - 4]
    
    return dp[n]    
````

````python
coin(5)
````

````python
coin(10)
````

### <a id='Ex2'> Ex.2 House Robber

You are a professional robber planning to rob houses along a street. Each house has a certain amount of money stashed, the only constraint stopping you from robbing each of them is that adjacent houses have security system connected andÂ it will automatically contact the police if two adjacent houses were broken into on the same night.

Given a list of non-negative integers representing the amount of money of each house, determine the maximum amount of money you can rob tonightÂ without alerting the police.

ä½ æ˜¯ä¸€ä¸ªå°å·ï¼Œç›¯ä¸Šä¸€æ’æˆ¿å­ï¼Œç›¸é‚»çš„æˆ¿å­è‹¥éƒ½è¢«æŠ¢å°†è§¦åŠ¨æŠ¥è­¦ç³»ç»Ÿï¼Œæ¯ä¸ªæˆ¿å­é‡Œé¢æœ‰å¤šå°‘é’±ä½ éƒ½è°ƒæŸ¥æ¸…æ¥šäº†ï¼Œ
æ±‚ä¸è§¦åŠ¨æŠ¥è­¦ç³»ç»Ÿçš„æƒ…å†µä¸‹æ€æ ·æ‰èƒ½å…¥å®¤æŠ¢åŠ«æŠ¢åˆ°æœ€å¤šé’±ï¼Ÿ

<a href="#top">å›åˆ°é¡¶éƒ¨</a>

````python
def rob(nums):
    n = len(nums)
    dp = [ [0] * 2 for _ in range(n + 1)]
    for i in range(1, n + 1):
        dp[i][0] = max(dp[i - 1][0], dp[i - 1][1]) # forget it
        dp[i][1] = nums[i - 1] + dp[i - 1][0]       # let's do it
    return max(dp[n][0], dp[n][1])  
````

````python
nums = [2,7,9,3,1]
rob(nums)
````

````python
def rob(nums):
    yes, no = 0, 0
    for i in nums: 
        no, yes = max(no, yes), i + no
    return max(no, yes)
````

````python
nums = [2,7,9,3,1]
rob(nums)
````

### <a id='Ex6'> Ex. 6 Min Cost Climbing Stairs æœ€å°å°é˜¶é—®é¢˜

On a staircase, theÂ i-th step has some non-negative costÂ cost[i]Â assigned (0 indexed).

Once you pay the cost, you can either climb one or two steps. You need to find minimum cost to reach the top of the floor, and you can either start from the step with index 0, or the step with index 1.

æœ‰ä¸€ä¸ªæ¥¼æ¢¯ï¼Œéœ€è¦ä½ ä»æœ€åº•ä¸‹åˆ°æœ€ä¸Šé¢ï¼Œä½ å¯ä»¥ä¸€æ¬¡èµ°ä¸€æ­¥æˆ–è€…ä¸€æ¬¡èµ°ä¸¤æ­¥<br>
é™åˆ¶ï¼šæ¯ä¸€å±‚æ¥¼æ¢¯éƒ½è¦äº¤è´¹ï¼Œæ¯å±‚æ”¶è´¹ä¸åŒ<br>
    
æ±‚ï¼šå¦‚ä½•èµ°æ‰èŠ±è´¹æœ€å°‘ï¼Ÿ<br>

<a href="#top">å›åˆ°é¡¶éƒ¨</a>

````python
# æ—¶é—´å¤æ‚åº¦ï¼šO(n), ç©ºé—´å¤æ‚åº¦ï¼šO(n)
def minCostClimbingStairs(cost):
    n = len(cost) + 1
    dp = [0] * n
    for i in range(2, n):
        dp[i] = min(dp[i - 2] + cost[i - 2], dp[i - 1] + cost[i - 1])
    return dp[-1]
````

````python
cost = [1, 100, 1, 1, 1, 100, 1, 1, 100, 1]
minCostClimbingStairs(cost)
````

````python
def minCostClimbingStairs2(cost):
    dp0, dp1, dp2 = 0, 0, 0
    for i in range(2, len(cost) + 1):
        dp2 = min(dp0 + cost[i - 2], dp1 + cost[i - 1])
        dp0, dp1 = dp1, dp2
    return dp2
````

````python
cost = [1, 100, 1, 1, 1, 100, 1, 1, 100, 1]
minCostClimbingStairs2(cost)
````

### <a id='Ex7'> Ex. 7 Decode Way  æ±‚ç¼–ç æ–¹å¼

A message containing letters fromÂ A-ZÂ is being encoded to numbers using the following mapping:<br>
ä¸€æ®µåŒ…å«ç€ A-Z çš„çŸ­ä¿¡ç”¨ä»¥ä¸‹æ–¹å¼è¿›è¡Œç¼–ç ï¼š
```
'A' -> 1 
'B' -> 2 
... 
'Z' -> 26 
```

Given an encoded message containing digits, determine the total number of ways to decode it.<br>
è¦æ±‚ï¼šç»™å®šä¸€æ®µç¼–ç çš„çŸ­ä¿¡ï¼Œè®¡ç®—ç¼–ç çš„æ–¹å¼ï¼š
```
Input: "12"
Output: 2
Explanation: It could be decoded as "AB" (1 2) or "L" (12).
```

```
Input: "226"
Output: 3
Explanation: It could be decoded as "BZ" (2 26), "VF" (22 6), or "BBF" (2 2 6).
```

<a href="#top">å›åˆ°é¡¶éƒ¨</a>

````python
def numDecodings(s):
    if s=="" or s[0]=='0': return 0
    dp=[1,1]
    for i in range(2,len(s)+1):
        # if it is 0, then dp[i] = 0
        result = 0
        if 10 <=int(s[i-2:i]) <=26:
            result = dp[i-2]
        if s[i-1] != '0':
            result += dp[i-1]
        dp.append(result)
    return dp[len(s)]
````

````python
numDecodings("110")
````

````python
numDecodings("40")
````

````python
numDecodings("226")
````

### <a id='Ex8'> Ex. 8 Unique Binary Search Trees ç‹¬ç‰¹äºŒå‰æ ‘çš„æœç´¢è·¯å¾„

GivenÂ n, how many structurally uniqueÂ BST'sÂ (binary search trees) that store values 1...n?<br>

ç»™å®š nï¼Œç”¨ 1, 2..., n æ¥è¡¨è¾¾ Binary Search Treeï¼Œæ±‚æœ‰å‡ ç§è¡¨è¾¾æ–¹å¼ï¼Ÿ


<a href="#top">å›åˆ°é¡¶éƒ¨</a>

````python
def numTress(n):
    if n <= 2:
        return n
    sol = [0] * (n + 1)
    sol[0] = sol[1] = 1
    for i in range(2, n + 1):
        for left in range (0, i):
            sol[i] += sol[left] * sol[i - 1 - left]
    
    return sol[n]        
````

````python
[numTress(i) for i in range(1, 6)]
````

### <a id='Ex9'> Ex. 9 Maximum Product Subarray æœ€å¤§å­åºåˆ—ä¹˜ç§¯

Find the contiguous subarray within an array (containing at least one number) which has the largest product.

ç»™ä½ ä¸€ä¸ªæ•°ç»„ï¼Œä½ éœ€è¦åœ¨æ•°ç»„ä¸­æ‰¾åˆ°ä¸€ä¸ªå­æ•°ç»„ï¼Œä½¿å…¶ä¹˜ç§¯ä¸ºæœ€å¤§

    
<a href="#top">å›åˆ°é¡¶éƒ¨</a>

````python
def maxProduct(nums):
    if len(nums) == 0:
        return 0
    maximum = minimum = result = nums[0]
    for i in range(1, len(nums)):
        maximum, minimum = max(maximum * nums[i], minimum * nums[i], nums[i]), \
                           min(maximum * nums[i], minimum * nums[i], nums[i])
        result = max(result, maximum)
    return result
````

````python
nums = [2,3,-2,4]
maxProduct(nums)
````
